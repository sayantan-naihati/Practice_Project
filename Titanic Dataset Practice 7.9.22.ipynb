```python
nullvalues=(df.isna().sum()/df.shape[0])*100
```

```python
nullvalues
```

```python
#Changing the Null values in Age column with mean values
```

```python
mean_value=df['Age'].mean()
```

```python
df['Age'].fillna(value=mean_value, inplace=True)
```

```python
df
```

```python
#We can also do the same by changing null values with median values...
```

```python
median_value=df['Age'].median()
```

```python
df['Age'].fillna(value=median_value, inplace=True)
```

```python
#Removing 2 rows from Dataset,which has null values in Embarked column...
```

```python
df=df.dropna(subset=['Embarked'])
```

```python
df.isnull().sum()
```

```python
#Replacing Null values from Cabin column by 'Unknown' class...
```

```python
df['Cabin'] = df['Cabin'].fillna("Unknown")
```

```python
df
```

```python
df.isnull().sum()
```

```python
#Checking for duplicate rows...
```

```python
df[df.duplicated()].sum().sum()
```

```python
#There is no duplicate row in the dataframe..
```

```python
#CATEGORICAL ENCODING
#Changing Sex Variable to numeric value...
#Male=1
#Female=0
```
